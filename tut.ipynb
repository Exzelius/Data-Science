```python
data = {'names':['steve','john','richard','sarah','randy','michael','julie'],
        'age':[20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}

ranking_df = DataFrame(data)
print(ranking_df)

ranking_df.iloc[2:5,1]=np.nan
ranking_df.iloc[3:6,3]=np.nan
ranking_df.iloc[3,:]=np.nan

ranking_df
```

```python
# Returns true on values which are missing
ranking_df.isnull() 
```

```python
# Returns true on values which are not missing
ranking_df.notnull()
```

```python
print(ranking_df['age'])
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]     
```

```python
# Grouping data by Index
address = 'mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.head()
```

```python
cars_groups = cars.groupby(cars['cyl'])
cars_groups.mean(numeric_only=True)         # Calculating means within the group of cyl
```
